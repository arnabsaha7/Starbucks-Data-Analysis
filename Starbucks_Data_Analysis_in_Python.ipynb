# Data Cleaning and Transformation

```python
## 3.2. Renaming Columns
df.rename(columns ={
    ' Total Fat (g)': 'total_fat_g',
    'Trans Fat (g) ': 'trans_fat_g',
    'Saturated Fat (g)': 'saturated_fat_g',
    ' Sodium (mg)': 'sodium_mg',
    ' Total Carbohydrates (g) ': 'total_carbohydrates_g',
    'Cholesterol (mg)': 'cholesterol_mg',
    ' Dietary Fibre (g)': 'dietary_fibre_g',
    ' Sugars (g)': 'sugar_g',
    ' Protein (g) ': 'protein_g',
    'Vitamin A (% DV) ': 'Vitamin_A_DV',
    'Vitamin C (% DV)': 'Vitamin_C_DV',
    ' Calcium (% DV) ': 'calcium_DV',
    'Iron (% DV) ': 'iron_DV',
    'Caffeine (mg)': 'caffeine_mg'}, inplace=True)
```

```python
# Converting Percentage Values

def func(val):
    if isinstance(val, str):
        val = val.replace('%', '')
    return float(val) / 100

df['Vitamin_A_DV'] = df['Vitamin_A_DV'].apply(func)
df['Vitamin_C_DV'] = df['Vitamin_C_DV'].apply(func)
df['calcium_DV'] = df['calcium_DV'].apply(func)
df['iron_DV'] = df['iron_DV'].apply(func)
```

```python
# Handling Missing Values and Data Type Conversion

df['total_fat_g'] = df['total_fat_g'].replace('3 2', '3.2')
df['total_fat_g'] = pd.to_numeric(df['total_fat_g'], errors='coerce')

df.drop_duplicates(inplace=True)

df['caffeine_mg'].fillna('125', inplace=True)

df = df[df['caffeine_mg'].str.lower() != 'varies']
df['caffeine_mg'] = pd.to_numeric(df['caffeine_mg'], errors='coerce')
```

# Data Visualization

```python
# Bar PLot for Beverage Categories

beverage_ctg_df = df['Beverage_category'].value_counts()

sns.barplot(x=beverage_ctg_df.index, y=beverage_ctg_df, palette='viridis')

plt.xlabel('Beverage Category')
plt.ylabel('Count')
plt.title('Beverage Category Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

> ‚ô¶  **This Graph shows the most sold drinks**

```python
# Bar PLot for Beverage Calories

beverage_ctg_df = df['Beverage_category'].value_counts()

sns.barplot(x=df['Beverage_category'], y=df['Calories'], palette='viridis')

plt.xlabel('Beverage Category')
plt.ylabel('Calories')
plt.title('Beverage Calories Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

> ‚ô¶  **This Graph shows the drinks with most Calories**

```python
# Doughnut Chart for Beverage Categories
plt.figure(figsize=(8, 18))
category_counts = df['Beverage_category'].value_counts()
category_counts.plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Distribution of Beverage Categories')
plt.ylabel('')

centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.show()
```

```python
# Box Plot for Key Nutritional Information
fig, axes = plt.subplots(3, 1, figsize=(6,9))

sns.boxplot(x=df['Calories'], ax= axes[0])
sns.boxplot(x=df['cholesterol_mg'], ax= axes[1])
sns.boxplot(x=df['total_fat_g'], ax= axes[2])

plt.tight_layout()
plt.show()
```

```python
# Filtering Data
x_calories = df.loc[(df.Calories > 500)]
x_total_fat = df.loc[(df.total_fat_g > 12)]
```

```python
# Caffeine Distribution
plt.figure(figsize=(10,5))
sns.histplot(df['caffeine_mg'], kde=True)
plt.title('Caffeine Distribution')
plt.xlabel('Caffeine')
plt.show()
```

```python
# Hist Plot for Total-Fat and Trans-Fat Distribution
plt.figure(figsize=(12,8))
sns.histplot(df['total_fat_g'], kde=True, label='Total Fat', color='violet', bins=20)
sns.histplot(df['trans_fat_g'], kde=True, label='Trans Fat', color='green', bins=20)

plt.xlabel(' Fat Content (g) ')
plt.ylabel('Frequency')
plt.title('Total Fat vs Trans Fat')
plt.legend()
plt.show()
```

```python
# Scatter Plot for Total Carbohydrates and Calories Distribution
plt.figure(figsize=(10,6))
sns.scatterplot(x='total_carbohydrates_g', y='Calories', hue='cholesterol_mg',
                palette='magma', data=df)
plt.xlabel("Total Carbs")
plt.ylabel('Calories')
plt.title('Carbs vs Cal')
plt.legend()
plt.show()
```

```python
# Correlation Matrix

numeric_df = df.select_dtypes(include='number')
corr_matrix = numeric_df.corr()

plt.figure(figsize=(10,10))
sns.heatmap(corr_matrix, annot=True, cmap="mako", fmt='.2f', linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```
